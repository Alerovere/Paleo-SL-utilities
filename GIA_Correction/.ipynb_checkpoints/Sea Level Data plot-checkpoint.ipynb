In this notebook, we calculate paleo Global Mean Sea Level (GMSL) starting from the following data:

1. a set of observed and dated relative sea-level (RSL) index points
2. an ensamble of Glacial Isostatic Adjustment (GIA) models 
3. an estimate of long-term tectonics and related uncertainties

After the data is imported, a monte-carlo approach is implemented (1 million repetitions), with the following calculations:

1. A random observation is chosen among those in the list
2. From that observation, the normal distributions of paleo RSL and age are sampled
3. One GIA correction is selected among the ensamble models, corresponding to the closest model time (no interpolation is done between model time steps). Only the GIA corrections corresponding to the interglacial are considered in this step (129-118 ka)
4. The tectonic rate and uncertainty are sampled within a normal distribution

At the end of the iterations, a dataframe is created containing the results. From the observed RSL, the GIA and tectonic contributions are then subtracted. Then, the results are plotted in a series of KDE plots showing the probabilistic distribution of RSL-only, GIA-corrected and Tectonics-corrected sea level.  

## Dependencies
For this script to work properly, the following dependencies have to be installed.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

## Import data
The data are imported from an .xlsx file for observations and one .csv file for GIA corrections. See examples in the project folder.

```python
#Import the RSL observations (xlsx file)
RSL=pd.read_excel('RSL_Observations.xlsx', header=0)

# Import the GIA model output (csv file)
GIA=pd.read_csv('GIA.csv', header=0)

#A new dataframe with clipped GIA is created,
#to maintain only the highstand corrections - 129 to 118 ka
GIA_clip=GIA[(GIA['Age']<=129) & (GIA['Age']>=118)].reset_index().drop(columns=['index'])
```

```python
RSL
```

## Tectonics correction
The numbers below should be changed according to available long-term tectonic rate values.

```python
rate=-0.04 # m/ka = mm/a
rate_unc=0.01 # m/ka = mm/a
```

## Plotting GIA and RSL
The code below plots the RSL data points and the ensamble of GIA corrections. These are the main inputs.

```python
# Modify the GIA dataframe for plotting with seaborn
GIA_plot = GIA.set_index('Age').stack().reset_index(name='RSL_pred').rename(columns={'level_1':'Model'})

# Creating plot
fig, ax = plt.subplots(nrows=1, ncols=2, sharey=True, squeeze=True,figsize=(15, 7))

# Define a palette with a single color, for the number of the ensambles (576 in the example)
palette = sns.color_palette(['black'], 576)

# Plot the RSL index points on the first graph.
rslplot=ax[0].errorbar(RSL['Age (ka)'], RSL['PRSL (m)'],
                     yerr=RSL['PRSL Error'], xerr=RSL['Age Error'], fmt='', 
                     ecolor='k',linestyle='')

# Plot the GIA on the second subplot 
giaplot=sns.lineplot(data=GIA_plot,x='Age',y='RSL_pred',
                   hue='Model',
                   legend=False,
                   palette=palette,
                   ax=ax[1],
                   alpha  = 0.05)

#Set x and y labels
ax[0].set_xlabel('Age (ka)', fontsize = 16)
ax[0].set_ylabel('RSL (m)', fontsize = 16)
giaplot.set_xlabel('Age (ka)', fontsize = 16)

# Set titles
ax[0].set_title('RSL index points', fontsize = 16)
giaplot.set_title('Background GIA correction', fontsize = 16)

# Axes limits are set manually, change the line below to adjust them
ax[0].set(xlim=(113,131),ylim=(-2,8))
ax[1].set(xlim=(113,131),ylim=(-2,8))

# Save current figure
plt.tight_layout()
fig.savefig('RSL_GIA_plot.jpg',dpi=600)
```

## Monte-carlo iteration
This cell performs the iteration described above. It may take a long time to execute. If the script crashes while executing, try reducing the number of iterations.

```python
########### CALCULATE PALEO RSL DATA ###########
Iterations=1000000

# Defining empty dataframes
time=[]
rsl=[]
gia=[]
tect=[]

#Creates a matrix randomly sampling the sea level data points and the GIA matrix
val = np.linspace(0, Iterations, num=Iterations+1)
for x in val:
 #Selects a random row
 rnd = RSL.sample(n=1)
 #Selects random time and RSL from normal distribution of age and RSL within the row
 t=np.random.normal(rnd['Age (ka)'], rnd['Age Error']/2, 1)
 time.append(float(t[0]))
 sl=np.random.normal(rnd['PRSL (m)'], rnd['PRSL Error'], 1)
 rsl.append(float(sl[0]))
 #Finds the closest GIA correction to the sampled time
 pos = GIA_clip['Age'].sub(float(t[0])).abs().values.argmin()
 #Selects a random GIA correction among those available for the sampled time
 GIArnd=np.random.randint(1, high=577, size=None, dtype=int)
 corr=GIA_clip.iloc[pos][GIArnd]
 gia.append(corr);
 #Samples the normal distribution of tectonic rates
 tec=np.random.normal(rate,rate_unc, 1)
 tect.append(float(t[0])*float(tec[0]))

#Creates the dataframe
RSLrand = pd.DataFrame({'Time':time, 'RSL':rsl,'GIA':gia,'Tect':tect})
RSLrand['Time'] = RSLrand['Time'].astype(float)
RSLrand['RSL'] = RSLrand['RSL'].astype(float)
RSLrand['GIA'] = RSLrand['GIA'].astype(float)
RSLrand['GIA_corrected'] = RSLrand['RSL']-RSLrand['GIA']
RSLrand['tect_corrected'] = RSLrand['GIA_corrected']-RSLrand['Tect']
```

```python
agemin=110
agemax=120
Max=RSLrand[(RSLrand['Time']<=agemax) & (RSLrand['Time']>=agemin)].reset_index().drop(columns=['index'])

Max['GIA_corrected'].quantile([.25,.5,.75])
```

##### RSL-only plot
This plot shows the age/elevation probabilistic distribution of RSL index points.

```python
# Plot the figure
graph2=sns.jointplot(x=RSLrand['Time'], y=RSLrand['GIA_corrected'], 
                     kind="kde", fill='True',
                     height=7, space=0,cmap='viridis')
graph2.set_axis_labels('Time [ka BP]', 'Corrected RSL [m]', fontsize=12)

#If needed, adjust the axes extents in the two lines below
graph2.ax_marg_x.set_xlim(110, 135)
graph2.ax_marg_y.set_ylim(-5, 8)

#Thanks: https://stackoverflow.com/questions/60358228/how-to-set-title-on-seaborn-jointplot
graph2.ax_joint.collections[0].set_alpha(0)
graph2.fig.tight_layout()
graph2.fig.subplots_adjust(top=0.95) # Reduce plot to make room 

plt.hlines(Max['GIA_corrected'].quantile([.25][0]), 0, 0.05)
plt.hlines(Max['GIA_corrected'].quantile([.5][0]), 0, 0.05)
plt.hlines(Max['GIA_corrected'].quantile([.75][0]), 0, 0.05)

#Save the figure
graph2.savefig("RSL_GIA_CORR_KDE.jpg",dpi=600)
```
